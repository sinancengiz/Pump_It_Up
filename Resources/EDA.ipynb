```python
# import train data
train_data = pd.read_csv('training_set_values.csv')
train_data_labels = pd.read_csv('training_set_labels.csv')
```

```python
# Put features and target together
train_data["status_group"] = train_data_labels["status_group"]
```

```python
# Convert status groop to numerical
# replace the functionalbut needs repair  with functional
train_data["status_group"] = train_data["status_group"].replace("functional" , 0 )
train_data["status_group"] = train_data["status_group"].replace("functional needs repair", 1)
train_data["status_group"] = train_data["status_group"].replace("non functional" , 2 )
```

```python
# Split validation set
train_set= train_data.iloc[:50000]
validation_set = train_data.iloc[50000:]
```

```python
# create validationset csv
validation_set.to_csv("validation_set.csv", index=False) 
train_data = train_set
```

```python
# this is ML file
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score , f1_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# seperate Labels and features
X =train_data.drop(["status_group"], axis=1) 
y = train_data["status_group"]
```

```python
# split data in train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)
#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

feature_imp = pd.Series(clf.feature_importances_,index=X_train.columns).sort_values(ascending=False)
print(feature_imp.head(30))
```

```python
feature_imp_top = feature_imp.head(30)
print(feature_imp_top.head(10))
```

```python
# Creating a bar plot
sns.barplot(x=feature_imp_top.index, y=feature_imp_top)
# Add labels to your graph
plt.xlabel('Features')
plt.ylabel('Feature Importance Score')
plt.title("Visualizing Important Features")
plt.legend()
plt.show()
```

```python
# create df by selected features
cleaned_train_data = train_data[feature_imp_top.index]
cleaned_train_data["status_group"] = train_data["status_group"]
```

```python
cleaned_train_data.to_csv("cleaned_train_data.csv", index=False) 
```
