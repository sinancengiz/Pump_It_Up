```python
# this is ML file
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import xgboost as xgb
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score , f1_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score

%matplotlib inline
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

```python
# import cleaned train data from csv
train_data = pd.read_csv('cleaned_train_data.csv')
```

```python
X = train_data.drop(["status_group"],axis = 1)
y = train_data["status_group"]
```

```python
# split data in train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y, test_size=0.20)
```

```python
# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler().fit(X_train)
```

```python
#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)
#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train_scaled,y_train)
feature_imp = pd.Series(clf.feature_importances_,index=X_train.columns).sort_values(ascending=False)
feature_imp_top = feature_imp.head(50)
print(feature_imp_top.head(10))
```

```python
X = X[feature_imp_top.index]
# split data in train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y, test_size=0.20)
```

```python
# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler().fit(X_train)

# Transform the training and testing data using the X_scaler and y_scaler models
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
rf_clf = RandomForestClassifier(random_state=42 , bootstrap =  True , n_estimators=1000,max_depth = 16, min_samples_split = 2 , max_features=  'auto', min_samples_leaf =  1)
rf_clf.fit(X_train_scaled, y_train)
print(rf_clf.feature_importances_)
y_pred = rf_clf.predict(X_test_scaled)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:", accuracy_score(y_test, y_pred))
print("F Score:",f1_score(y_test,y_pred, average='weighted'))
```

```python
# Combined report with all above metrics
from sklearn.metrics import classification_report
print(classification_report(y_test , y_pred ))
```

```python
validation_set = pd.read_csv('validation_set.csv')
test_data = validation_set.drop("status_group",axis=1)
test_labels = validation_set["status_group"]
# test_data = pd.read_csv('test_set_values.csv')
```

```python
test_data.describe()
```

```python
# Binned lat long columns
test_data['bined_long'] = np.where(test_data['longitude'] >= 1, 'yes', 'no')
test_data['bined_lat'] = np.where(test_data['latitude'] >= -1, 'no', 'yes')
len(test_data[test_data["bined_long"] == "no"])
```

```python
dummy_df = test_data
dummy_df_not_0 = dummy_df[dummy_df["longitude"] > 2]
dummy_df = dummy_df_not_0.groupby(['district_code']).mean()
dummy_df = dummy_df.reset_index()
dummy_df = dummy_df[["district_code","longitude","latitude"]]
dummy_df = dummy_df.rename(columns={ "longitude": "longitude_temp", "latitude": "latitude_temp"})

test_data = test_data.join(dummy_df.set_index('district_code'), on='district_code')

longitude = []
for index , row in test_data.iterrows():
    if row["longitude"] < 5:
        longitude.append(row["longitude_temp"])
    else:
        longitude.append(row["longitude"])

test_data["longitude"] = longitude

latitude = []
for index, row in test_data.iterrows():
    if row["latitude"] > -1:
        latitude.append(row["latitude_temp"])
    else:
        latitude.append(row["latitude"])
        
test_data["latitude"] = latitude

test_data = test_data.drop(["longitude_temp", "latitude_temp"] , axis = 1)
```

```python
# filling  0 value gps_heigt rows with mean value of thier geografic region
dummy_df = test_data
dummy_df_not_0 = dummy_df[dummy_df["gps_height"] >= 1]
dummy_df = dummy_df_not_0.groupby(["district_code"]).mean()
dummy_df = dummy_df.reset_index()
dummy_df = dummy_df[["district_code", "gps_height"]]
dummy_df = dummy_df.rename(columns={ "gps_height": "gps_height_temp"})

test_data = test_data.join(dummy_df.set_index("district_code"), on='district_code')

gps_height = []
for index , row in test_data.iterrows():
    if row["gps_height"] <= 1:
        gps_height.append(row["gps_height_temp"])
    else:
        gps_height.append(row["gps_height"])

test_data["gps_height"] = gps_height

test_data["gps_height"] = test_data["gps_height"].fillna(test_data["gps_height"].mean())

test_data = test_data.drop(["gps_height_temp"] , axis = 1)
```

```python
# getting 5 fisrt value can reduce unique values to 6 column to get dummy variables later
dummy_list = test_data["funder"]
emty_dummy_list = []
first_5_value = ['Government Of Tanzania', 'Danida','Hesawa', 'World Bank', 'Rwssp']


for value in dummy_list:
    if value in  first_5_value:
        emty_dummy_list.append(value)
    else:
        emty_dummy_list.append("Others")

test_data["funder"] = emty_dummy_list
```

```python
# getting 5 fisrt value can reduce unique values to 6 column to get dummy variables later
dummy_list = test_data["installer"]
emty_dummy_list = []
first_5_value = ['DWE', 'Government','RWE','Commu','DANIDA']

for value in dummy_list:
    if value in  first_5_value:
        emty_dummy_list.append(value)
    else:
        emty_dummy_list.append("Others")

test_data["installer"] = emty_dummy_list
```

```python
# Stacking with 2 model 
model_1 = RandomForestClassifier(random_state=42 , bootstrap =  True , n_estimators=1000,max_depth = 16, min_samples_split = 2 , max_features=  'auto', min_samples_leaf =  1)
model_3 = KNeighborsClassifier(n_neighbors=7)
model_2 = xgb.XGBClassifier(random_state=42,  n_estimators=1000 , max_depth = 16, min_child_weight = 3  , gamma = 0 , 
                            subsample = 0.8, colsample_bytree = 0.8 , scale_pos_weight = 1 ,learning_rate =0.1)

model_1.fit(X_train_scaled, y_train)
model_2.fit(X_train_scaled, y_train)
model_3.fit(X_train_scaled, y_train)

pred_1 = model_1.predict(X_test_scaled)
pred_2 = model_2.predict(X_test_scaled)
pred_3 = model_3.predict(X_test_scaled)

test_pred_1 = model_1.predict(test_data_scaled)
test_pred_2 = model_2.predict(test_data_scaled)
test_pred_3 = model_3.predict(test_data_scaled)

stacked_pred = np.column_stack((pred_1, pred_2, pred_3))
stacked_test_pred = np.column_stack((test_pred_1, test_pred_2 , test_pred_3))

meta_model= SVC()
meta_model.fit(stacked_pred , y_test)
```
